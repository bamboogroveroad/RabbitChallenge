```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import cluster, preprocessing, datasets

from sklearn.cluster import KMeans
```

```python
wine = datasets.load_wine()
```

```python
print(wine.DESCR)
```

```python
X = wine.data
```

```python
X.shape
```

```python
y=wine.target
```

```python
y.shape
```

```python
wine.target_names
```

```python
model = KMeans(n_clusters=3)
```

```python
labels = model.fit_predict(X)
```

```python
df = pd.DataFrame({'labels': labels})
type(df)
```

```python
def species_label(theta):
    if theta == 0:
        return wine.target_names[0]
    if theta == 1:
        return wine.target_names[1]
    if theta == 2:
        return wine.target_names[2]
```

```python
df['species'] = [species_label(theta) for theta in wine.target]
```

```python
pd.crosstab(df['labels'], df['species'])
```

上記の結果、各クラスへの分類があまりうまくいっていない。そこでデータを標準化してから再度実行する。

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

sc = StandardScaler()
sc.fit(X)

sn = sc.transform(X)
result = model.fit(sn)
df = pd.DataFrame({'labels': result.labels_})

def species_label(theta):
    if theta == 0:
        return wine.target_names[0]
    if theta == 1:
        return wine.target_names[1]
    if theta == 2:
        return wine.target_names[2]

df['species'] = [species_label(theta) for theta in wine.target]
pd.crosstab(df['labels'], df['species'])
```

⇒かなり改善した。
